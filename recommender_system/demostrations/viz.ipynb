```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the LaTeX table/TikZ figure
categories = ["Binary (No Read)", "Binary (Read)", "Weighted (No Read)", "Weighted (Read)"]
# Values for MAP@K and NDCG@K based on the provided coordinates
map_k = [0.00184, 0.00177, 0.00526, 0.00505]
ndcg_k = [0.00396, 0.00132, 0.01588, 0.00928]

# Set up the positions and width for the bars
x = np.arange(len(categories))  # x locations for the groups
width = 0.35  # width of each bar

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Create the bars for each metric
rects1 = ax.bar(x - width/2, map_k, width, label='MAP@K')
rects2 = ax.bar(x + width/2, ndcg_k, width, label='NDCG@K')

# Add labels, title, and custom x-axis tick labels
ax.set_ylabel('Scores')
ax.set_title('Comparison of MAP@K and NDCG@K scores')
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45, ha='right')
ax.set_ylim(0, 0.02)
ax.legend(loc='upper left')

# Function to add data labels above each bar
def autolabel(rects):
    """Attach a text label above each bar displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.5f}',
                    xy=(rect.get_x() + rect.get_width()/2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

# Adjust layout and display the plot
fig.tight_layout()
plt.show()
```
